```python
# load data
import pandas as pd

df = pd.read_csv('Salary_Data.csv')
df[:2]
```

```python
# 1 input: YearsExperience
# 1 Output: Salary
# check relationship of X and y
df.plot(kind='scatter', x='YearsExperience', y='Salary')
```

```python
# prepare input and output
df[:2]
```

```python
X = df[['YearsExperience']]  # input
y = df[['Salary']] # output
```

```python
# Problem is Simple linear Regression
# load algorithm
from sklearn.linear_model import LinearRegression
```

```python
# sklearn is main ML library which contains all functionalities required to build ML model
#from sklearn---> algorithms
```

```python
# lets build a model
model = LinearRegression()
model
```

```python
# training phase using fit() method
model.fit(X,y)
```

```python
# testing using predict() method
model.predict([[0]])
```

```python
model.predict([[2],[3.5],[30]])
```
